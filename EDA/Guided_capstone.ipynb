```python
import statsmodels.formula.api as smf
results= smf.ols('AdultWeekday ~ total_chairs + Runs + vertical_drop+ summit_elev+ fastQuads', data=df).fit()
results.params
```

```python
df2 = pd.DataFrame()
df2['Runs']= np.linspace(3,341)
df2['vertical_drop']= np.linspace(60,4425)
df2['summit_elev']= np.linspace(315,13487)
df2['fastQuads']= np.linspace(0,15)
df2['total_chairs']= 14
p12= results.predict(df2)
```

Using ordinary least squares, it can be seen that ticket prices for resorts with total chair lifts >=14 is > $86. This increment in current prices will be enough to offset additional operational cost.
